```python
import re 
import pandas as pd
```

```python
f=open("WhatsApp Chat with Gear_up.txt",encoding='utf-8')
data=f.read()
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
messages = re.split(pattern, data)[1:]
len(messages)
```

```python
dates = re.findall(pattern, data)
len(dates)
print(dates[1:4])
```

```python
df=pd.DataFrame({'u_messages':messages,'m_dates':dates})
#convert msg datatype
df['m_dates']=pd.to_datetime(df['m_dates'], format='%d/%m/%Y, %H:%M - ')
```
